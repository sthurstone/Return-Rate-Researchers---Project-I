```python
test_df = pd.DataFrame(r['data'])
test_df.head()
```

```python
df_e = test_df['entities'].apply(pd.Series)
df_e
```

```python
df_a=df_e['annotations'].apply(pd.Series)
df_a
```

```python
df_a_0=df_a[0].apply(pd.Series)
df_a_0
```

```python
ann2_df = df_a_0[["normalized_text","probability","type"]]
ann2_df
```

```python
df_h=df_e['hashtags'].apply(pd.Series)
df_h
```

```python
df_h_0=df_h[0].apply(pd.Series)
df_h_0
```

```python
h2_df=df_h_0[["tag"]]
h2_df
```

```python
df_u=df_e['urls'].apply(pd.Series)
df_u
```

```python
df_u_0=df_u[0].apply(pd.Series)
df_u_0
```

```python
url_df=df_u_0[["title","url"]]
url_df
```

```python
df_m=df_e['mentions'].apply(pd.Series)
df_m
```

```python
df_m_0=df_m[0].apply(pd.Series)
df_m_1=df_m[1].apply(pd.Series)
df_m_total=pd.concat([df_m_0,df_m_1],axis=1)
df_m_total
```

```python
df_m_total2=df_m_total.fillna(0)
df_m_total2
```

```python
mtotal_df=df_m_total2[["username"]]
mtotal_df
```

```python
mtotal_df.columns=['username 1', 'username 2']
mtotal_df
```

```python
mtotal_df.dtypes
```

```python
mtotal_df['temp 1'] = [1 if x else 0 for x in mtotal_df['username 1']]
mtotal_df['temp 2'] = [1 if x else 0 for x in mtotal_df['username 2']]
mtotal_df['Mentions Total'] = mtotal_df["temp 1"] + mtotal_df["temp 2"]
mtotal_df
```

```python
mtotal2_df = mtotal_df.drop(['temp 1', 'temp 2'], axis=1)
mtotal2_df
```

```python
public_metrics_df = pd.json_normalize(test_df['public_metrics'])
public_metrics_df.head()
```

```python
public_metrics_df.columns=['retweet count', 'reply count', 'like count', 'quote count']
public_metrics_df
```

```python
#public_metrics1_df = public_metrics_df[(public_metrics_df['retweet count']>=1)&(public_metrics_df['like count']>=1)]
#public_metrics1_df
```

```python
df_new = test_df.join(ann2_df)
df_new2 = df_new.join(h2_df)
df_new3 = df_new2.join(url_df)
df_new4 = df_new3.join(mtotal2_df)
merged_df = df_new4.join(public_metrics_df)
merged_df
```

```python
merged2_df = merged_df.drop(['entities', 'public_metrics', 'username 1', 'username 2'], axis=1)
merged2_df
```

```python
merged3_df = merged2_df[(merged2_df['lang']=='en')]
merged3_df
```
