```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import glob

# Study data files
Facebook_Ads = "Facebook_Ads_2.csv"
KAG_Conversions= "KAG_conversion_data.csv"
Social_Network_Ads="Social_Network_Ads(1).csv"
Social_Influence="social-influence-on-shopping.csv"

# Read the mouse data and the study results
Facebook_Ads = pd.read_csv(Facebook_Ads)
KAG_Conversions = pd.read_csv(KAG_Conversions)
Social_Network_Ads=pd.read_csv(Social_Network_Ads)
Social_Influence=pd.read_csv(Social_Influence)

#Add blank column to merge data
Facebook_Ads["Column"] = ""
KAG_Conversions["Column"] = ""
Social_Network_Ads["Column"] = ""
Social_Influence["Column"]=""
```
